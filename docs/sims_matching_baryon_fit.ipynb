```python
from multiprocessing import Pool
import pathos.pools as pp
import numpy as np
import astropy.stats
import supercubos
import corner
from sklearn.gaussian_process.kernels import Matern
import emcee
import maszcal.lensing
import maszcal.emulate
import maszcal.interpolate
import maszcal.likelihoods
import maszcal.fitutils
import maszcal.data.sims
import maszcal.data.obs
from maszcal.interp_utils import cartesian_prod
```

```python
LOWER_RADIUS_CUT = 0.125
UPPER_RADIUS_CUT = 3
```

```python
sim_data = maszcal.data.sims.NBatta2010('../data/NBatta2010/').cut_radii(LOWER_RADIUS_CUT, UPPER_RADIUS_CUT)
```

```python
cov_magnitude = 1.3
cov = np.identity(sim_data.radii.size) * cov_magnitude
fisher = np.identity(sim_data.radii.size) / cov_magnitude
sim_errs = np.sqrt(np.diag(cov))
print(sim_errs)
```

```python
masses = sim_data.masses
zs = np.repeat(sim_data.redshifts, masses.shape[1])
masses = masses.flatten()

density = maszcal.density.MatchingGnfw(
    cosmo_params=sim_data.cosmology,
    mass_definition='mean',
    delta=500,
    units=None,
    comoving_radii=True,
    nfw_class=maszcal.density.MatchingNfwModel,
)

model = maszcal.lensing.MatchingShearModel(
    sz_masses=masses,
    redshifts=zs,
    lensing_weights=np.ones(zs.size),
    rho_func=density.rho_tot,
)

rs = sim_data.radii
```

```python
param_mins = np.array([1, 0.1, 0.1, -2]) #con, alpha, beta, a_sz
param_maxes = np.array([6, 6, 8.1, 2])

lh = supercubos.LatinSampler().get_sym_sample(param_mins, param_maxes, 800)
```

```python
def _pool_map(func, array):
    pool = pp.ProcessPool()
    mapped_array = np.array(
        pool.map(func, array),
    ).T
    pool.close()
    pool.join()
    pool.terminate()
    pool.restart()
    return mapped_array
```

```python
def esd_func(params):
    con = params[0:1]
    alpha = params[1:2]
    beta = params[2:3]
    gamma = np.array([0.2])
    a_sz = params[3:4]
    return np.squeeze(model.stacked_delta_sigma(rs, a_sz, con, alpha, beta, gamma))
```

```python
esds = _pool_map(esd_func, lh)
```

```python
esds.shape
```

```python
data = rs[:, None] * esds

emulator = maszcal.emulate.PcaEmulator.create_from_data(
    coords=lh,
    data=data,
    interpolator_class=maszcal.interpolate.GaussianProcessInterpolator,
    interpolator_kwargs={'kernel': Matern()}
)
```

```python
rand = supercubos.LatinSampler().get_rand_sample(
    param_mins=param_mins,
    param_maxes=param_maxes,
    num_samples=800,
)
```

```python
esds_check = _pool_map(esd_func, rand)
data_check = rs[:, None] * esds_check
```

```python
emulated_data = emulator(rand)
errs = (emulated_data - data_check)/data_check

colors = ['#72a555', '#ab62c0', '#c57c3c', '#638ccc', '#ca5670']

plt.plot(rs, 100*np.abs(errs).mean(axis=-1), color=colors[0], linestyle='--')
plt.ylim((0, 3))
plt.xscale('log')
plt.xlabel(r'$r (Mpc)$')
plt.gcf().set_size_inches(6, 3)
pass
```

```python
plt.hist(np.log10(np.abs(errs.mean(axis=0))), color=colors[4], bins=50)
plt.xlabel(r'$\log_{10}(\mathrm{emulation \, error})$')
plt.ylabel('number of test profiles')
plt.title('800 sample Latin Hypercube in 4D')
plt.gcf().set_size_inches(4, 3)
plt.savefig('../figs/act-talk/emulator.pdf')
```

```python
def log_like(params, data):
    model = emulator(params[None, :]).flatten()
    return maszcal.likelihoods.log_gaussian_shape(model, data, fisher)


def log_prob(params, data):
    if np.all(param_mins < params) and np.all(params < param_maxes):
        return log_like(params, data)
    else:
        return - np.inf
```

```python
plt.plot(rs, rs[:, None]*sim_data.wl_signals.reshape(24, -1), color=colors[3], alpha=0.05,
         zorder=1)
sim_stack = rs * (sim_data.wl_signals.mean(axis=(1, 2)))
plt.errorbar(rs, sim_stack, yerr=sim_errs, color=colors[4], capsize=1, capthick=0.5)
plt.xscale('log')
plt.gcf().set_size_inches(6, 4)
pass
```

```python
best_fit = maszcal.fitutils.global_minimize(
    lambda p: -log_like(p, sim_stack),
    param_mins,
    param_maxes,
    method='global-differential-evolution',
)
```

```python
best_fit
```

```python
import emcee

nwalkers, ndim = (500, 4)
pos = best_fit + 0.6e-4 * np.random.randn(nwalkers, ndim)
nsteps = 4000

with Pool() as pool:
    sampler = emcee.EnsembleSampler(nwalkers, ndim, log_prob, args=(sim_stack,), pool=pool)
    sampler.run_mcmc(pos, nsteps, progress=True);
```

```python
param_names = [r'$c$', r'$\alpha$', r'$\beta$', r'$a_{sz}$']
colors = ['#5b99e2', '#48c7a9']
fig,ax = plt.subplots(4, sharex=True)
for i in range(4):
    ax[i].plot(sampler.chain[:, :, i].T, color=colors[1], alpha=0.2)
    ax[i].set_title(param_names[i])
```
